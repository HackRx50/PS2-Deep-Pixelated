```python
# Install necessary packages
!pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118

!pip install fastapi uvicorn nest_asyncio pyngrok paddleocr transformers pillow ipywidgets


# Import statements
import os
import cv2
import torch
import time
import numpy as np
from paddleocr import PaddleOCR
from transformers import Qwen2VLForConditionalGeneration, AutoProcessor
from PIL import Image
from io import BytesIO
from fastapi import FastAPI, UploadFile, File
from fastapi.responses import JSONResponse
import uvicorn
import nest_asyncio
import threading
from pyngrok import ngrok
from ipywidgets import HTML
from IPython.display import display

# Apply nest_asyncio
nest_asyncio.apply()

# Set the default event loop policy
import asyncio
asyncio.set_event_loop_policy(asyncio.DefaultEventLoopPolicy())

# Initialize FastAPI app
app = FastAPI()

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Initialize OCR
ocr_device = 'gpu' if torch.cuda.is_available() else 'cpu'
ocr = PaddleOCR(use_angle_cls=True, lang='en', device=ocr_device)

# Initialize Qwen2VLForConditionalGeneration model and processor
model = Qwen2VLForConditionalGeneration.from_pretrained(
    "Qwen/Qwen2-VL-2B-Instruct", torch_dtype="auto", device_map="auto"
).to(device)
processor = AutoProcessor.from_pretrained("Qwen/Qwen2-VL-2B-Instruct")

# Define utility functions
def read_image_from_bytes(image_bytes):
    image = Image.open(BytesIO(image_bytes)).convert("RGB")
    return image

def crop_provisional_diagnosis(img, ocr_results):
    for line in ocr_results:
        for word_info in line:
            text = word_info[1][0].lower()
            if "provisional diagnosis" in text:
                bbox = word_info[0]
                x_min = int(min([point[0] for point in bbox]))
                y_min = int(min([point[1] for point in bbox])) - 30
                x_max = int(max([point[0] for point in bbox])) + 1000
                y_max = int(max([point[1] for point in bbox])) + 100

                cropped_img = img[y_min:y_max, x_min:x_max]
                return cropped_img
    return None

def extract_provisional_diagnosis(image_bytes):
    start_time = time.time()
    # Read image from bytes
    img = cv2.imdecode(np.frombuffer(image_bytes, np.uint8), cv2.IMREAD_COLOR)
    # Use OCR to get results
    result = ocr.ocr(img, cls=True)
    cropped_img = crop_provisional_diagnosis(img, result)
    elapsed_time = time.time() - start_time
    print(f"Time taken for extract_provisional_diagnosis: {elapsed_time:.2f} seconds")
    return cropped_img

def process_vision_info(messages):
    image_inputs = []
    for message in messages:
        for content in message["content"]:
            if content["type"] == "image":
                # content["image"] is a PIL Image
                image_inputs.append(content["image"])
    return image_inputs, None

def extract_text_from_image(image):
    start_time = time.time()

    messages = [
        {
            "role": "user",
            "content": [
                {"type": "image", "image": image},
                {"type": "text", "text": "Extract provisional diagnosis value from the image"}
            ],
        }
    ]

    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    image_inputs, _ = process_vision_info(messages)
    inputs = processor(text=[text], images=image_inputs, padding=True, return_tensors="pt").to(device)
    generated_ids = model.generate(**inputs, max_new_tokens=128)
    generated_ids_trimmed = [
        out_ids[len(in_ids):] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
    ]
    output_text = processor.batch_decode(
        generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
    )
    elapsed_time = time.time() - start_time
    print(f"Time taken for extract_text_from_image: {elapsed_time:.2f} seconds")
    return output_text[0] if output_text else ""

def refine_extracted_text(extracted_text):
    start_time = time.time()

    messages = [
        {
            "role": "user",
            "content": [
                {"type": "text", "text": extracted_text},
                {"type": "text", "text": "Refine the extracted diagnosis. Correct spelling mistakes and ensure that any misrecognized medical terms are corrected based on context."}
            ],
        }
    ]

    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = processor(text=[text], padding=True, return_tensors="pt").to(device)
    generated_ids = model.generate(**inputs, max_new_tokens=128)
    generated_ids_trimmed = [
        out_ids[len(in_ids):] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
    ]
    refined_text = processor.batch_decode(
        generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
    )
    elapsed_time = time.time() - start_time
    print(f"Time taken for refine_extracted_text: {elapsed_time:.2f} seconds")
    return refined_text[0] if refined_text else ""

# Define the endpoint
@app.post("/extract_diagnosis")
async def extract_diagnosis(file: UploadFile = File(...)):
    try:
        # Read the uploaded file
        image_bytes = await file.read()
        # Extract provisional diagnosis section
        cropped_img = extract_provisional_diagnosis(image_bytes)
        if cropped_img is not None:
            # Convert cropped_img (numpy array) to PIL Image
            cropped_image_pil = Image.fromarray(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
            # Extract text from cropped image
            extracted_diagnosis = extract_text_from_image(cropped_image_pil)
            # Refine the extracted diagnosis
            refined_diagnosis = refine_extracted_text(extracted_diagnosis)
            # Return the results
            return {
                "extracted_diagnosis": extracted_diagnosis,
                "refined_diagnosis": refined_diagnosis
            }
        else:
            return JSONResponse(content={"error": "Could not extract provisional diagnosis section"}, status_code=400)
    except Exception as e:
        return JSONResponse(content={"error": str(e)}, status_code=500)

# Start the server in a new thread with loop="asyncio" and use_uvloop=False
def run():
    uvicorn.run(app, host='0.0.0.0', port=8000, loop="asyncio")

threading.Thread(target=run, daemon=True).start()

# Set up the ngrok tunnel
ngrok_tunnel = ngrok.connect(8000)
print(f"Public URL: {ngrok_tunnel.public_url}")

# Display server status
t = HTML(
    value="Server is running...",
    description='',
    style={'description_width': 'initial'},
)
display(t)
```
